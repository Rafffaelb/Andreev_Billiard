```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
df_Concurrence_C_N1 = pd.read_csv(r'Concurrence/Andreev_Concurrence_C_Gamma_N1.txt', sep='\t', header = None)
df_Concurrence_CI_N1 = pd.read_csv(r'Concurrence/Andreev_Concurrence_CI_Gamma_N1.txt', sep='\t', header = None)

df_Entanglement_C_N1 = pd.read_csv(r'Concurrence/Andreev_Entanglement_C_Gamma_N1.txt', sep='\t', header = None)
df_Entanglement_CI_N1 = pd.read_csv(r'Concurrence/Andreev_Entanglement_CI_Gamma_N1.txt', sep='\t', header = None)
```

```python
df_Concurrence_C_N1.dropna(inplace = True)
df_Concurrence_CI_N1.dropna(inplace = True)
df_Concurrence_C_N1.rename(columns = lambda s: (s/20), inplace = True)
df_Concurrence_C_N1.head()
```

```python
df_Concurrence_CI_N1.rename(columns = lambda s: (s/20), inplace = True)
df_Concurrence_C_N1.head()
```

```python
df_Entanglement_C_N1.dropna(inplace = True)
df_Entanglement_CI_N1.dropna(inplace = True)
df_Entanglement_C_N1.rename(columns = lambda s: (s/20), inplace = True)
df_Entanglement_C_N1.head()
```

```python
Concurrence_C_N1 = df_Concurrence_C_N1.values
Concurrence_CI_N1 = df_Concurrence_CI_N1.values

Entanglement_C_N1 = df_Entanglement_C_N1.values
Entanglement_CI_N1 = df_Entanglement_CI_N1.values
```

```python
fig, axs = plt.subplots(2,4, figsize = (14,6), constrained_layout = True)

fig.suptitle(r'DISTRIBUTION OF CONCURRENCE AS FUNCTION OF $\Gamma$ (N = 1)', fontsize = 20)

for j in range(0,4):
    
    gamma_idx = 5*j+5
    
    gamma_value = 0.25*gamma_idx/5
    
    axs[0,j].hist(Concurrence_C_N1[:,gamma_idx], bins = 30, color='orange', label = r'$C$', alpha=1, density = True, histtype='bar', ec='black');
    axs[1,j].hist(Concurrence_CI_N1[:,gamma_idx], bins = 30, color='dodgerblue', label = r'$CI$', alpha=1, density = True, histtype='bar', ec='black');
    
    axs[0,j].set_title(r'$\Gamma = $'+ str(gamma_value), fontsize = 18)
    
    axs[0,0].set_ylabel(r'$\mathcal{P}_{C}(\mathcal{C})$', fontsize = 18)
    axs[1,0].set_ylabel(r'$\mathcal{P}_{CI}(\mathcal{C})$', fontsize = 18)
    
        
    for i in range(0,2):
        axs[i,0].legend(fontsize = 20)
        axs[i,j].grid(linestyle=':')
        axs[i,j].tick_params(axis = 'both', which = 'major', labelsize = 12)
```

```python
df_Concurrence_C_N2 = pd.read_csv(r'Concurrence/Andreev_Concurrence_C_Gamma_N2.txt', sep='\t', header = None)
df_Concurrence_CI_N2 = pd.read_csv(r'Concurrence/Andreev_Concurrence_CI_Gamma_N2.txt', sep='\t', header = None)

df_Entanglement_C_N2 = pd.read_csv(r'Concurrence/Andreev_Entanglement_C_Gamma_N2.txt', sep='\t', header = None)
df_Entanglement_CI_N2 = pd.read_csv(r'Concurrence/Andreev_Entanglement_CI_Gamma_N2.txt', sep='\t', header = None)
```

```python
df_Concurrence_C_N2.dropna(inplace = True)
df_Concurrence_C_N2.rename(columns = lambda s: (s/20), inplace = True)
df_Concurrence_C_N2.head()
```

```python
df_Concurrence_CI_N2.dropna(inplace = True)
df_Concurrence_CI_N2.rename(columns = lambda s: (s/20), inplace = True)
df_Concurrence_CI_N2.head()
```

```python
Concurrence_C_N2 = df_Concurrence_C_N2.values
Concurrence_CI_N2 = df_Concurrence_CI_N2.values

Entanglement_C_N2 = df_Entanglement_C_N2.values
Entanglement_CI_N2 = df_Entanglement_CI_N2.values
```

```python
def Concurrence_distribution_C(x):
    for i in range(len(x)):
        p_C_x[i] = (5/8)*((x[i]**3)/((1-(x[i]**2))**6))*(2*np.arctanh(np.sqrt(1-(x[i]**2)))*(560+5880*(x[i]**2)+7350*(x[i]**4)+1225*(x[i]**6))-(2904+16264*(x[i]**2)+10350*(x[i]**4)+512*(x[i]**6))*np.sqrt(1-(x[i]**2)))
    return p_C_x

def Concurrence_distribution_CI(x):
    for i in range(len(x)):
        p_CI_x[i] = 15*((x[i]*(1-x[i]))/(np.sqrt(1-x[i]**2)*((1+x[i])**3)))
    return p_CI_x

fig, axs = plt.subplots(2,4, figsize = (14,6), constrained_layout = True)

fig.suptitle(r'DISTRIBUTION OF CONCURRENCE AS FUNCTION OF $\Gamma$ (N = 2)', fontsize = 20)

for j in range(0,4):
    
    gamma_idx = 5*j+5
    
    gamma_value = 0.25*gamma_idx/5
    
    axs[0,j].hist(Concurrence_C_N2[:,gamma_idx], bins = 30, color='orange', label = r'$C$', alpha=1, density = True, histtype='bar', ec='black');
    axs[1,j].hist(Concurrence_CI_N2[:,gamma_idx], bins = 30, color='dodgerblue', label = r'$CI$', alpha=1, density = True, histtype='bar', ec='black');
    
    axs[0,j].set_title(r'$\Gamma = $'+ str(gamma_value), fontsize = 18)
    
    axs[0,0].set_ylabel(r'$\mathcal{P}_{C}(\mathcal{C})$', fontsize = 18)
    axs[1,0].set_ylabel(r'$\mathcal{P}_{CI}(\mathcal{C})$', fontsize = 18)
    
    if (j == 3):
        x = [i/100 for i in range(0,100)]
        p_C_x = [[] for i in range(0,100)]
        p_CI_x = [[] for i in range(0,100)]
        p_C_x = Concurrence_distribution_C(x)
        p_CI_x = Concurrence_distribution_CI(x)
        axs[0,j].plot(x, p_C_x, color = 'black', label = r'$\mathcal{P}_{C}(\mathcal{C})$', linestyle = 'dashed', linewidth = 2)
        axs[1,j].plot(x, p_CI_x, color = 'black', label = r'$\mathcal{P}_{2}(\mathcal{C})$', linestyle = 'dashed', linewidth = 2)
        axs[0,3].legend(fontsize = 15, loc = 'upper right')
        axs[1,3].legend(fontsize = 15, loc = 'upper right')
    
    for i in range(0,2):
        axs[i,0].legend(fontsize = 20)
        axs[i,j].grid(linestyle=':')
        axs[i,j].tick_params(axis = 'both', which = 'major', labelsize = 12)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

X_data = np.array([i/20 for i in range(0,21)])
X = X_data.reshape(-1,1)

y_data = (meanConcurrence_O-meanConcurrence_U)
y = y_data.reshape(-1,1)

degree=2
polyreg = make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg.fit(X, y)

X_seq = np.linspace(X.min(),X.max(),300).reshape(-1,1)

print("\n######################################\n")

print("Coefficients of the concurrence polynomial (degree = 2):")
print(polyreg.steps[1][1].coef_)
print("\nIntercept of the concurrence polynomial (degree = 2):")
print(polyreg.steps[1][1].intercept_)

fig, axs = plt.subplots(1,2, figsize = (15,6), constrained_layout = True, facecolor = 'w')

fig.suptitle(r'QIC OF CONCURRENCE $\mathcal{C}^{QIC}$ AND ENTANGLEMENT OF FORMATION $\mathcal{E}^{QIC}$ AS FUNCTION OF $\Gamma$ FOR ANDREEV BILLIARD', fontsize = 20)

axs[0].scatter(X,y, s = 100, color = 'purple', label = 'Simulation')
axs[0].plot(X_seq,polyreg.predict(X_seq),color="black", linestyle = 'dashed', label = 'Quadratic regression')
axs[0].grid(linestyle=':')
axs[0].set_xlabel(r'$\Gamma$', fontsize = 21)
axs[0].tick_params(axis = 'both', which = 'major', labelsize = 18)
axs[0].set_ylabel(r'$\mathcal{C}^{QIC}$', fontsize = 25)
axs[0].legend(fontsize = 20)

X_data = np.array([i/20 for i in range(0,21)])
X = X_data.reshape(-1,1)

y_data = (meanEntanglement_O-meanEntanglement_U)
y = y_data.reshape(-1,1)

degree=2
polyreg=make_pipeline(PolynomialFeatures(degree),LinearRegression())
polyreg.fit(X, y)

print("\n######################################\n")
print("Coefficients of the entanglement of formation polynomial (degree = 2):")
print(polyreg.steps[1][1].coef_)
print("\nIntercept of the entanglement of formation polynomial (degree = 2):")
print(polyreg.steps[1][1].intercept_)

X_seq = np.linspace(X.min(),X.max(),300).reshape(-1,1)
axs[1].scatter(X,y, s = 100 ,color = 'purple')
axs[1].plot(X_seq,polyreg.predict(X_seq),color="black", linestyle = 'dashed')
axs[1].grid(linestyle=':')
axs[1].set_xlabel(r'$\Gamma$', fontsize = 21)
axs[1].tick_params(axis = 'both', which = 'major', labelsize = 18)
axs[1].set_ylabel(r'$\mathcal{E}^{QIC}$', fontsize = 25)
```
