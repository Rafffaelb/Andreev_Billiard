```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
df_Bell_Parameter_Ress_C = pd.read_csv(r'Bell_Parameter/Bell_Ress/Andreev_Bell_Parameter_C_Ress.txt', sep = '\t', header = None)
df_Bell_Parameter_Ress_CI = pd.read_csv(r'Bell_Parameter/Bell_Ress/Andreev_Bell_Parameter_CI_Ress.txt', sep = '\t', header = None)
```

```python
df_Bell_Parameter_Ress_C.rename(columns = lambda s: ((s*25)+50))
```

```python
df_Bell_Parameter_Ress_CI.rename(columns = lambda s: ((s*25)+50))
```

```python
Bell_Parameter_Ress_C = df_Bell_Parameter_Ress_C.values
Bell_Parameter_Ress_CI = df_Bell_Parameter_Ress_CI.values

print("VIOLATIONS OF BELL PARAMETER WITHOUT TRS ENSEMBLE")
print(df_Bell_Parameter_Ress_C[df_Bell_Parameter_Ress_C > 2.0].count())
print("\n#############################################\n")
print("VIOLATIONS OF BELL PARAMETER WITH TRS ENSEMBLE")
print(df_Bell_Parameter_Ress_CI[df_Bell_Parameter_Ress_CI > 2.0].count())

Violations_C = df_Bell_Parameter_Ress_C[df_Bell_Parameter_Ress_C > 2.0].count().values
Violations_CI = df_Bell_Parameter_Ress_CI[df_Bell_Parameter_Ress_CI > 2.0].count().values
```

```python
print("MAX VALUES OF BELL PARAMETER WITHOUT TRS ENSEMBLE")
print(df_Bell_Parameter_Ress_C.max())
print("\n")
print("MAX VALUES OF BELL PARAMETER WITH TRS ENSEMBLE")
print(df_Bell_Parameter_Ress_CI.max())
```

```python
fig, axs = plt.subplots(1,2, figsize = (14, 6), constrained_layout = True)

fig.suptitle(r'HISTOGRAM OF $P(|\mathcal{S}_{CHSH}|)$ FOR ANDREEV BILLIARD WITH 300 RESONANCES', fontsize = 20)

axs[0].hist(Bell_Parameter_Ress_CI[:,0], bins = 50, color = 'dodgerblue', label = r'TRS', alpha = 1, density = True, histtype = 'bar', ec = 'black')
axs[1].hist(Bell_Parameter_Ress_C[:,10], bins = 50, color = 'red', label = r'No TRS', alpha = 1, density = True, histtype = 'bar', ec = 'black')

for i in range(2):
    axs[i].legend(prop = {'size': 20}, bbox_to_anchor = (0.7, 0.97))
    axs[i].axvline(2, color = 'black')
    axs[i].grid(linestyle = ':')
    axs[i].set_ylabel('$P(|\mathcal{S}_{CHSH}|)$', fontsize = 20)
    axs[i].tick_params(axis = 'both', which = 'major', labelsize = 20)
    axs[i].set_xlabel(r'$|\mathcal{S}_{CHSH}|$', fontsize = 20)
```

```python
fig, axs = plt.subplots(1,1, figsize = (8, 6), constrained_layout = True, facecolor = 'w')


axs.scatter([25*i + 50 for i in range(0,11)], (100/1000000)*Violations_CI, label = 'TRS', s = 100, color = 'dodgerblue', alpha = 1)
axs.scatter([25*i + 50 for i in range(0,11)], (100/1000000)*Violations_C, label = 'No TRS', s = 100, color = 'red', alpha = 1)
    
axs.plot([25*i + 50 for i in range(0,11)], (100/1000000)*Violations_CI, color = 'dodgerblue', linestyle = 'dashed')
axs.plot([25*i + 50 for i in range(0,11)], (100/1000000)*Violations_C, color = 'red', linestyle = 'dashed')

fig.suptitle(r'VIOLATIONS IN TERMS OF RESONANCES FOR ANDREEV BILLIARD', fontsize = 20)
axs.set_xlabel('Resonances', fontsize = 15)
axs.set_ylabel('Violations (%)', fontsize = 19)
axs.set_ylim([0, 0.8])
axs.tick_params(axis = 'both', which = 'major', labelsize = 15)
axs.grid(linestyle = ':')
axs.legend(fontsize = 15)
```
